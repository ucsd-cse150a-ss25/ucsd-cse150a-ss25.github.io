```python
import networkx as nx
```

## Initialize our [Graph](https://networkx.org/documentation/stable/reference/classes/graph.html#networkx.Graph)

```python
G = nx.Graph()
```

```python
# add a node with value 1
G.add_node(1)
```

```python
# Let's print our degree
G.degree()
```

```python
print(G)
```

```python
G.add_nodes_from([2, 3])
```

```python
print(G)
```

```python
# add some edges
G.add_edge(1, 2)
# unpack edge tuple*
e = (2, 3)
G.add_edge(*e)
```

## Printing our Graph

```python
import matplotlib.pyplot as plt
```

### Plotting using [newtorkx.draw](https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw.html) and [matplotlib](https://matplotlib.org/stable/api/index.html)
More on [networkx drawing](https://networkx.org/documentation/stable/reference/drawing.html)

```python
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, with_labels=True, font_weight='bold')
```

## Let's make some more graphs using lists

```python
G.add_nodes_from([1, 2, 3, 4, 5])
```

```python
G.add_edges_from([(1, 2), (1, 3), (3, 4), (2, 5)])
```

```python
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, with_labels=True, font_weight='bold')
```

## Let's make a new graph using dictionaries

```python
G.add_nodes_from([(1, {"color": "blue"}), (2, {"color": "sky blue"}), (3, {"color": "pink"}), (4, {"color": "red"}), (5, {"color": "green"})])
```

```python
G.add_edges_from([(1, 2), (2, 3), (3, 4), (2, 5), (3,5)])
```

```python
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, with_labels=True, font_weight='bold')
```
